```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist = range(360)
dist2 = [1]*360
```

```python
plt.plot(dist)
```

```python
scatter_dist = np.random.choice(range(360), size=(10000), p=np.array(dist)/sum(dist))
scatter_dist2 = np.random.choice(range(360), size=(10000), p=np.array(dist2)/sum(dist2))
```

```python
plt.hist(scatter_dist)
```

```python
plt.hist(scatter_dist2)
```

```python
arr = np.ndarray((2, 10000), dtype=np.uint16)
arr[0, :] = scatter_dist
arr[1, :] = scatter_dist2
```

```python
plt.hist(arr[0, :])
```

```python
plt.hist(arr[1, :])
```
